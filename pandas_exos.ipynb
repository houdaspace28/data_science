```python
import numpy as np
import pandas as pd
df= pd.read_csv('titanic.csv')
print(df.dtypes)
print(df.describe())
```

```python
df_copy = df.copy()
dummies = pd.get_dummies(df_copy['name'])
print(dummies)
```

```python
age_above_30 = df[df['age'] > 30]
print(age_above_30)

mean_age = df.groupby('sex')['age'].mean()
print(mean_age)

df['adult'] = np.where(df['age']>18, 'Adult', 'Child')
print(df['adult'])
```

```python
pivot_1 = pd.pivot_table(df,values='age',index='pclass',aggfunc='mean')
pivot_1

crosstab_1 = pd.crosstab(df['sex'], df['survived'])
crosstab_1
```
