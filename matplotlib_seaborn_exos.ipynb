```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print('importing done')
```

```python
data = {
    'Year': range(2000, 2024),
    'Population': [1000000 * (1.02 ** i) + np.random.normal(0, 10000) for i in range(24)],
    'Growth_Rate': [2 + np.random.normal(0, 0.5) for _ in range(24)],
    'Birth_Rate': [15 + np.random.normal(0, 1) for _ in range(24)],
    'Death_Rate': [8 + np.random.normal(0, 0.5) for _ in range(24)],
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 24)
}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 6))
plt.plot(df['Year'],df['Population'])
plt.title('Population growth over time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
data_2=np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(8,8))
plt.hist(data_2, bins=30, edgecolor='black')
plt.title('Histogram of data')
plt.xlabel('Data')
plt.ylabel('Frequency')     
plt.show()
```

```python

# Read data
df = pd.read_csv('iris.csv')

# Basic scatter plot with custom colors
plt.figure(figsize=(10, 6))

# Method 1: Using different colors for each species
colors = {
    'setosa': 'red',
    'versicolor': 'blue',
    'virginica': 'green'
}

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], 
                subset['sepal_width'],
                c=colors[species],        # Color specification
                label=species,            # For legend
                alpha=0.6)                # Transparency

plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
box = plt.boxplot(df['sepal_length'])
plt.title('Box Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')

# Add annotations to explain each part
plt.annotate('Maximum (excluding outliers)', 
            xy=(1, box['whiskers'][1].get_ydata()[1]),
            xytext=(1.2, box['whiskers'][1].get_ydata()[1]),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('75th percentile (Q3)', 
            xy=(1, box['boxes'][0].get_ydata()[2]),
            xytext=(1.2, box['boxes'][0].get_ydata()[2]),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Median', 
            xy=(1, box['medians'][0].get_ydata()[0]),
            xytext=(1.2, box['medians'][0].get_ydata()[0]),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('25th percentile (Q1)', 
            xy=(1, box['boxes'][0].get_ydata()[0]),
            xytext=(1.2, box['boxes'][0].get_ydata()[0]),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Minimum (excluding outliers)', 
            xy=(1, box['whiskers'][0].get_ydata()[1]),
            xytext=(1.2, box['whiskers'][0].get_ydata()[1]),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()

# Print actual values
print("\nStatistical Summary:")
print(df['sepal_length'].describe())
```
